# Visualize the relationship between age and height

## Step 1: Import the dataset Howell1 into a pandas dataframe
https://github.com/rmcelreath/rethinking/blob/master/data/Howell1.csv

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set values for variables for age for children cutoff and colors for gender
age_cutoff_children = 17
color_female = 'green'
color_male = 'black'
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rmcelreath/rethinking/master/data/Howell1.csv', sep=';')
```

```python
df
```

## Step 2: Extract the list of ages and heights from the dataframe into 2 lists (age_list and height_list)

```python
age_list = df["age"].tolist()
height_list = df["height"].tolist()
```

```python
print(age_list)
```

```python
print(height_list)
```

## Step 3: Graph the scatterplot

```python
plt.scatter(age_list, height_list)
plt.show()
```

## Step 4: Filter out everyone over the age of a cutoff age to separate children from adults

```python
filtered_df_children = df[df['age'] < age_cutoff_children] 
```

```python
age_list_filtered = filtered_df_children["age"].tolist()
height_list_filtered = filtered_df_children["height"].tolist()
```

```python
plt.scatter(age_list_filtered, height_list_filtered)
plt.show()
```

```python
col = filtered_df_children.male.map({0:color_female, 1:color_male})
filtered_df_children.plot.scatter(x='age', y='height', c=col)
plt.show()
```

```python
filtered_df_adult = df[df['age'] >= age_cutoff_children]
age_list_filtered_adult = filtered_df_adult["age"].tolist()
height_list_filtered_adult = filtered_df_adult["height"].tolist()
```

```python
plt.scatter(age_list_filtered_adult, height_list_filtered_adult)
plt.show()
```

```python
col = filtered_df_adult.male.map({0:color_female, 1:color_male})
filtered_df_adult.plot.scatter(x='age', y='height', c=col)
plt.legend(["male", "female"])
```

```python
def calculate_mean_height(df):
    return df["height"].mean()
calculate_mean_height(df)
```

```python
def calculate_mean_height_by_age(df):
    return df.groupby("age")["height"].mean()

calculate_mean_height_by_age(df)
```

```python
#visualizations of the mean height by age
calculate_mean_height_by_age(df).plot()
plt.show()
```

```python
#visualize individual data points  
plt.scatter(age_list, height_list)
plt.show()
```

```python
#graph bar chart of mean height by age
calculate_mean_height_by_age(df).plot.bar()
plt.show()
```

```python
# graph children and adults separately
calculate_mean_height_by_age(filtered_df_children).plot.bar()
plt.show()
#graph adults
calculate_mean_height_by_age(filtered_df_adult).plot.bar()
plt.show()
```
